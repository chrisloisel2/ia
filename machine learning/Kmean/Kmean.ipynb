```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

### Affichage de iris avec seaborn

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
plt.title('Iris Dataset')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(iris.data[:, 2], iris.data[:, 3], c='black', s=100)
plt.show()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data
```

## Cration d'un graphique de dispersion de iris en utilisant seaborn

```python
import seaborn as sns

sns.scatterplot(data=data, x='sepal length (cm)', y='sepal width (cm)')
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)

kmeans.fit(data)
```

```python
data['cluster'] = kmeans.labels_
```

```python
print("centroids: ", kmeans.cluster_centers_)
print("inertia: ", kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'], c=data['cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.title('Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
